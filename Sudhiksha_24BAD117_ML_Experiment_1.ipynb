```python
#SCENARIO 1
#SUDHIKSHA 24BAD117
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("data1.csv", encoding='ISO-8859-1')

print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
print(df.isnull().sum())

sales_per_product = df.groupby("Description")["Quantity"].sum().head(10)

# Bar chart
sales_per_product.plot(kind='bar')
plt.title("Top 10 Products by Sales Quantity")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.show()

# Line chart
sales_per_product.plot(kind='line')
plt.title("Sales Trend of Top Products")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.show()
```

```python
# SCENARIO 2: HOSPITAL DATA
#SUDHIKSHA 24BAD117

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("diabetes.csv")

print(df.head())
print(df.info())
print(df.isnull().sum())

# Histograms
df["Glucose"].hist()
plt.title("Glucose Level Distribution")
plt.show()

df["Age"].hist()
plt.title("Age Distribution")
plt.show()

# Boxplot
df.boxplot(column="Glucose")
plt.title("Glucose Boxplot")
plt.show()

df.boxplot(column="Age")
plt.title("Age Boxplot")
plt.show()
```

```python
# SUDHIKSHA 24BAD117
# SCENARIO 3: HOUSING DATA

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("housing.csv")

# Display columns and missing values
print(df.columns)
print(df.isnull().sum())

# Scatter plot: Area vs Price
plt.scatter(df["area"], df["price"])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()

# Correlation matrix (numeric columns only)
corr = df.select_dtypes(include='number').corr()

# Heatmap using matplotlib
plt.figure(figsize=(10, 6))
plt.imshow(corr, interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# SUDHIKSHA 24BAD117
# SCENARIO 4: CUSTOMER MARKETING DATA ANALYSIS

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("marketing_campaign.csv")
print("Dataset Columns:")
print(df.columns)

print("\nDataset Info:")
print(df.info())

print("\nNull Values:")
print(df.isnull().sum())

df["Age"] = 2024 - df["Year_Birth"]


# Bar Plot  Age Distribution
plt.figure(figsize=(10,5))
df["Age"].value_counts().sort_index().plot(kind="bar")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.title("Age Distribution of Customers")
plt.show()

# Box Plot  Income Distribution
plt.figure(figsize=(6,4))
plt.boxplot(df["Income"].dropna())
plt.xlabel("Income")
plt.title("Income Distribution of Customers")
plt.show()

# Box Plot  Total Spending Pattern
df["Total_Spending"] = (
    df["MntWines"] +
    df["MntFruits"] +
    df["MntMeatProducts"] +
    df["MntFishProducts"] +
    df["MntSweetProducts"] +
    df["MntGoldProds"]
)

plt.figure(figsize=(6,4))
plt.boxplot(df["Total_Spending"])
plt.xlabel("Total Spending")
plt.title("Customer Spending Pattern")
plt.show()
```
